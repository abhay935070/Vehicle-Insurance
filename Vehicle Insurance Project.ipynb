# --> Check the outlier use of Boxplot

```python
sns.boxplot(df.Annual_Premium,orient='h')
plt.show()
```

```python
Q1,Q3 = df.Annual_Premium.quantile([0.25,0.75])
Q1,Q3
```

```python
IQR = Q3-Q1
IQR
```

```python
UL = Q3 + 1.75 * (IQR)
LL = Q1 - 1.75 * (IQR)
UL,LL
```

```python
df[df.Annual_Premium>UL]
```

```python
df['Annual_Premium'] = np.where(df.Annual_Premium>UL,UL,df.Annual_Premium)
```

```python
sns.boxplot(df.Annual_Premium,orient='h')
```

```python
sns.boxplot(df.Vintage,orient='h')
plt.show()
```

# --> Changing the column type for better visualization

```python
df['Gender'] = df['Gender'].replace({'M': 'Male', 'F': 'Female'})
df['Vehicle_Damage'] = df['Vehicle_Damage'].map({'Yes': 1, 'No': 0})
df['Vehicle_Age'] = df['Vehicle_Age'].map({'< 1 Year': 0, '1-2 Year': 1, '> 2 Years': 2})
```
